### 1. Indexing

Index rows of dataframes using slices [a:b], `.loc()` [label-based] and `.iloc()` [position-based].

````python
times_df.head(5)
````

````python
times_df.loc[2]
````

````python
times_df.iloc[2]  # in this case, labels and positions are identical so .loc() returns same as .iloc().
````

````python
times_df.loc[0, 'total_score']  # total score of the first row
````

````python
times_df[0:2]  # rows 0 and 1
````

````python
times_df.iloc[1:4, 1:4] # columns and rows 1-3
````

````python
times_df.total_score  # the whole total_score column
````

````python
# Are the last ten entries after 2006?
shanghai_df.iloc[-10:,'year'] > 2006  # can't do it the way he suggests, causes error
````

Can't do it the way he suggests, causes error because `.iloc[]` requires integer arguments.  
Otherwise, will work if you supply the integer position for 'year'.

````python
shanghai_df.iloc[-10:, 10] > 2006
````

OK, let's try using `.loc.` then.

````python
shanghai_df.loc[-10:, 'year']  
````

Can't do it this way either as `.loc` requires label references and -10 is not one of the labels.  
However, since positive integers are, something like `shanghai_df.loc[0:10, 'year']` will work but that's not the question here. (See next example)

````python
shanghai_df.loc[shanghai_df.index[-10:,], 'year'] > 2006  # this is the way to do it if you must use .loc
````

````python
# Was the alumni score higher than 90 for the top 10 universities?
shanghai_df.loc[0:10, 'alumni'] > 90
````

Can use `query()` to perform queries on the DataFrame

````python
# Query shanghai_df for schools with total_score between 40-50 
average_schools = shanghai_df.query('total_score >= 40 and total_score <= 50')
average_schools.sort_values('total_score', 0, False) # quicker to supply positional arguments axis and ascending
                                                   # than to skip axis and type keyword ascending = False
````

Lots of repeats, let's just look at one year.

````python
shanghai_df.groupby('year').count()
````

2015 is most recent and has as much data as any other year

````python
average_schools = shanghai_df.query('total_score >= 40 and total_score < 50 and year == 2015')
average_schools.sort_values('total_score', 0, False) 
````

````python
average_schools.max()  # careful, some of the data is str
````

## Method chaining with pipe()

````python
# define extract_info
def extract_info(input_df, name):
    df = input_df.copy()
    info_df = pd.DataFrame({'nb_rows': df.shape[0], 'nb_cols': df.shape[1], 'name': name}, index=range(1))
    return info_df

all_info = pd.concat([times_df.pipe(extract_info, 'times'), shanghai_df.pipe(extract_info, 'shanghai')])
````

````python
all_info
````

OK, maybe not the best concept to demonstrate pipe().  Usually I think of two or three nested pipes in a row like `h(g(f(x))) -> f(x).pipe(g).pipe(h)` or however the syntax goes.
